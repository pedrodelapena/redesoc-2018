#### <i> "A good hub page is one that points to many good authorities; a good authority page is one that is pointed to by many good hub pages."</i>

```python
#clonando nodes caso ocorra cactus
gtmc = gtm.copy()

#adicionando colunas 
for i in gtmc.nodes:
    gtmc.nodes[i]['indegree'] = 0
    gtmc.nodes[i]['outdegree'] = 0
    gtmc.nodes[i]['outdegreeNorm'] = 0
    gtmc.nodes[i]['indegreeNorm'] = 0
    gtmc.nodes[i]['indegreeC'] = 0
    gtmc.nodes[i]['nomes'] = gtmc.nodes[i]['label'] #just for A E S T H E T I C purposes
    
#indegree predecessor - outdegree successor
```

```python
total = 0
norm = 0
dictHub = {}
dictAut = {}

#total de votos de um genero musical (indegree)
for i in gtmc.nodes:
    counter = 0
    for j in gtmc.predecessors(i):
        counter += 1
    gtmc.nodes[i]['indegree'] = counter
    #print(gtmc.nodes[i]['indegree'], gtmc.nodes[i]['nomes'])
    
#"pontuação" dos hubs com base no numero de votos que um estilo que ele(a) votou recebeu (outdegree)
for i in gtmc.nodes:
    counter = 0
    for j in gtmc.successors(i):
        counter += gtmc.nodes[j]['indegree']
    gtmc.nodes[i]['outdegree'] = counter
    #print(gtmc.nodes[i]['outdegree'], gtmc.nodes[i]['nomes'])

#calculo do indegree dos estilos com base no outdegree das pessoas (bons hubs -> boas autoridades)
for i in gtmc.nodes:
    counter = 0
    for j in gtmc.predecessors(i):
        counter += gtmc.nodes[j]['outdegree']    
    gtmc.nodes[i]['indegreeC'] = counter
    total += counter
    #print(gtmc.nodes[i]['indegreeC'], gtmc.nodes[i]['nomes'])

#normalizando pontuações (authorities)
for i in gtmc.nodes:
    if gtmc.nodes[i]['indegreeC'] != 0:
        gtmc.nodes[i]['indegreeNorm'] = gtmc.nodes[i]['indegreeC']/total
        dictAut.update({gtmc.nodes[i]['indegreeNorm']: gtmc.nodes[i]['nomes']})
        norm += gtmc.nodes[i]['indegreeNorm']
        #print(gtmc.nodes[i]['indegreeNorm'], gtmc.nodes[i]['nomes'])
        
#normalizando pontuações (hubs)
for i in gtmc.nodes:
    counter = 0
    if gtmc.nodes[i]['outdegree'] != 0:
        for j in gtmc.successors(i):
            counter += gtmc.nodes[j]['indegreeNorm']
        gtmc.nodes[i]['outdegreeNorm'] = counter
        dictHub.update({gtmc.nodes[i]['outdegreeNorm']: gtmc.nodes[i]['nomes']})
        #print(gtmc.nodes[i]['outdegreeNorm'], gtmc.nodes[i]['nomes'])
    
#Resultados!
autSort = sorted(dictAut.items(), key=lambda kv: kv[0], reverse=True)
hubSort = sorted(dictHub.items(), key=lambda kv: kv[0], reverse=True)

print("Melhores hubs (teoricamente são os 'melhores' DJs)")
for i in range(len(hubSort)):
    print(hubSort[i])

print("")

print("Melhores autoridades (teoricamente são os 'melhores' gêneros musicais)")
for i in range(len(autSort)):
    print(autSort[i])
    
##Implementar terceira métrica (da uma lida no aviso do blackboard)
```
